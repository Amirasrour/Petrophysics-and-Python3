import libraries 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import lasio
```

```python
df = lasio.read('15-9-19_SR_COMP.LAS.txt').df()
```

```python
df.head(20)
```

```python
df.reset_index(inplace =True)
df.head(10)
```

```python
df.describe()
```

```python
df.dropna(inplace= True)
df.describe()
```

Creating scatter plot

```python
plt.scatter(x='NEU', y='DEN', data = df)
plt.xlim(-5, 60)
plt.ylim(3,1.5)
plt.xlabel('Neutron Porosity(NEU)-%', fontsize=14)
plt.ylabel('Bulk Density(DEN)-%', fontsize=14)
plt.show()
```

Changing Colormap and Adding Color bar by adding third variable 

```python
fig = plt.figure(figsize=(8,8) )
plt.scatter(x='NEU', y='DEN', data = df, c='GR',vmin=0, vmax=100)
plt.xlim(-5, 60)
plt.ylim(3,1.5)
plt.xlabel('Neutron Porosity(NEU)-%', fontsize=14)
plt.ylabel('Bulk Density(DEN)-%', fontsize=14)
plt.show()
```

```python
fig = plt.figure(figsize=(8,8) )
plt.scatter(x='NEU', y='DEN', data = df, c='GR',vmin=0, vmax=100, cmap= 'rainbow')
plt.colorbar(label=' Gamma Ray- API')
plt.xlim(-5, 60)
plt.ylim(3,1.5)
plt.xlabel('Neutron Porosity(NEU)-%', fontsize=14)
plt.ylabel('Bulk Density(DEN)-%', fontsize=14)
plt.show()
```

Adding Gridlines & Plot Styling

```python
plt.style.use('bmh')
fig = plt.figure(figsize=(8,8) )
plt.scatter(x='NEU', y='DEN', data = df, c='GR',vmin=0, vmax=100, cmap= 'rainbow')
plt.colorbar(label=' Gamma Ray- API')
plt.xlim(-5, 60)
plt.ylim(3,1.5)
plt.xlabel('Neutron Porosity(NEU)-%', fontsize=14)
plt.ylabel('Bulk Density(DEN)-%', fontsize=14)
plt.show()
```
